# **Numpy**

NumPy (Numerical Python) is the foundational library that performs lightning-fast math operations on massive datasets. Behind the scenes, almost every other ML library (including Pandas and Scikit-Learn) relies on NumPy arrays.

## **The CoreConcept: The N-Dimensional Array *(ndarray)***



*   Unlike standard Python lists, NumPy arrays require all data to be of the same type (usually numbers).

*   This single-type requirement makes them incredibly fast and memory-efficient.




## **Vectorization & Broadcasting**



*   This is the most critical NumPy concept for ML.
*   You do not use ***for*** loops in NumPy.
*   Instead, you apply mathematical operations directly to the entire array at once, which is known as Vectorization.

### **Phase-2: The Missing Essentials**

#### **Matrix Multiplication *(np.dot(A,B))***

The absolute core of deep learning.

```python
import numpy as np

features = np.array([2,5]) # e.g., bedrooms, age of house
weights = np.array([100, -5]) # e.g., value per bedroom, penalty per year

# Multiply features by weights to get a price prediciton
prediction = np.dot(features, weights)
prediction
```

#### **Randomization *(np.random.seed(), np.random.rand())***

Used to randomly shuffle your training data (so the model doesn't memorize the order) or to initialize starting weights for a neural network.

```python
import numpy as np

# Locking the seed ensures we get the same "random" numbers every time
np.random.seed(42)

# Generate 3 random weights between 0 and 1
random_weights = np.random.rand(3)

random_weights
```

**What is *np.random.seed(42)***

Computers cannot generate truly random numbers; they use math formulas to create "pseudo-random" numbers.

A seed is the starting point for that math formula.

By explicitly setting the seed to 42, you are telling the computer: "Start your random sequence from this exact specific point."

Why this matters for ML: When you build a model and it gets 95% accuracy, you want to be able to run the code again tomorrow and get the exact same result. If you don't lock the seed, your initial weights will change every time you run the script, making it impossible to debug or share your work reliably. (Note: The number 42 is just a long-standing programmer joke from 'The Hitchhiker's Guide to the Galaxy'. You could use 1, 100, or 999—as long as you keep it consistent).

**What is *np.random.rand(3)***

you are calling the rand() function from NumPy's random module.

The (3) tells NumPy the shape of the array you want. In this case, a 1-Dimensional array containing 3 items.

The rand() function specifically generates numbers strictly between 0 and 1.

Why this matters for ML: Neural networks start as a blank slate. Before they can learn, their connections (weights) need to be initialized with small, random numbers so the math algorithms have a starting point to optimize from.

#### **Memory Management *(array_name.astype())***

Converting data from heavy 64-bit floats down to 32-bit floats is a standard trick to train models faster and save memory.

```python
import numpy as np
heavy_data = np.array([1.1,2.2,3.3])
print(heavy_data.dtype)   # Output: float64 (Uses 64 bits per number)
print(heavy_data)

# Compressing to 32 bits to save RAM
light_data = heavy_data.astype('float32')

print(light_data.dtype)   # Output: float32
print(light_data)
```
